# Сортировки

**Python встроенные сортировки**


*   sorted()
*   sort()



Функция **sorted()** — это встроенный метод Python для сортировки элементов итерируемых объектов. Она создает новый отсортированный список, не изменяя исходный объект.

```python
sorted(iterable, key=None, reverse=False)
```
Параметры функции:


*   **iterable** — обязательный параметр, принимающий любой итерируемый объект (список, кортеж, строку, множество)
*   **key** — необязательный параметр, функция для определения порядка сортировки
*   **reverse** — необязательный параметр, определяет порядок сортировки (False — по возрастанию, True — по убыванию)



````python
# Базовая сортировка

numbers = [4, 1, 3, 2]
sorted_numbers = sorted(numbers)
print(sorted_numbers)
````

````python
# Сортировка в обратном порядке

numbers = [5, 2, 9, 1, 5, 6]
sorted_desc = sorted(numbers, reverse=True)
print(sorted_desc)
````

````python
# Сортировка строк

words = ["apple", "banana", "cherry", "date"]
sorted_words = sorted(words)
print(sorted_words)
````

````python
# Сортировка по длине строки

words = ["apple", "banana", "cherry", "date"]
sorted_by_length = sorted(words, key=len)
print(sorted_by_length)
````

````python
# Сортировка списка словарей

data = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 91},
    {"name": "Eve", "score": 78}
]

sorted_data = sorted(data, key=lambda x: x['score'])
print(sorted_data)
````

Особенности работы:


*   Функция работает с любыми итерируемыми объектами
*   Создает новый список, оставляя исходный объект неизменным
*   По умолчанию сортировка выполняется по возрастанию
*   Поддерживает пользовательские функции для определения порядка сортировки



Метод **sort()** — это встроенный метод списков в Python, который сортирует элементы непосредственно в исходном списке. В отличие от функции sorted(), он не создает новый список, а модифицирует существующий.

```python
list.sort(key=None, reverse=False)
```

Основные параметры:


*   **key** — функция для определения ключа сортировки
*   **reverse** — логическое значение (True/False) для сортировки в обратном порядке



````python
# Простая сортировка чисел

numbers = [5, 2, 9, 1, 5, 6]
numbers.sort()
print(numbers)
````

````python
# Сортировка строк

words = ["banana", "apple", "cherry"]
words.sort()
print(words)
````

````python
# Сортировка по убыванию

numbers = [5, 2, 9, 1, 5, 6]
numbers.sort(reverse=True)
print(numbers)
````

````python
# Сортировка по длине строк

words = ["banana", "apple", "cherry"]
words.sort(key=len)
print(words)
````

````python
# Сортировка списка словарей

students = [
    {"name": "John", "age": 25},
    {"name": "Jane", "age": 22},
    {"name": "Dave", "age": 23}
]

# Сортировка по возрасту
students.sort(key=lambda x: x["age"])
print(students)
````

````python
# Сортировка кортежей

matrix = [(1, 2), (3, 1), (2, 3)]
matrix.sort(key=lambda x: x[1])
print(matrix)
````

# Алгоритмы сортировки



*   **Пузырьковая сортировка** — простой алгоритм последовательного сравнения элементов
*   **Сортировка слиянием** — эффективный алгоритм с разделением массива
*   **Быстрая сортировка** — рекурсивный алгоритм с выбором опорного элемента

**Пузырьковая сортировка** — это алгоритм сортировки, который последовательно сравнивает соседние элементы списка и меняет их местами, если они расположены в неправильном порядке. Название происходит от аналогии с пузырьками в воде, которые поднимаются на поверхность.

**Принцип работы**
Алгоритм работает следующим образом:


1.   Сравниваются первые два элемента списка
2.   Если они находятся в неправильном порядке — меняются местами
3.   Процесс повторяется для всех пар элементов
4.   После каждого полного прохода самый большой элемент “всплывает” в конец списка

````python
def bubble_sort(arr):
    n = len(arr)
    # Внешний цикл для прохода по всем элементам
    for i in range(n):
        # Флаг для отслеживания перестановок
        swapped = False
        # Внутренний цикл для сравнения соседних элементов
        for j in range(0, n-i-1):
            # Если текущий элемент больше следующего
            if arr[j] > arr[j+1]:
                # Меняем их местами
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        # Если перестановок не было, массив уже отсортирован
        if not swapped:
            break

# Пример использования
numbers = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(numbers)
print("Отсортированный массив:", numbers)
````

Преимущества:


*   Простота реализации
*   Не требует дополнительной памяти
*   Хорошо работает на небольших массивах

Недостатки:

*   Низкая эффективность на больших массивах
*   Не подходит для практического применения в реальных задачах
*   Медленнее большинства других алгоритмов сортировки


**Сортировка слиянием** — это эффективный алгоритм сортировки, основанный на принципе «разделяй и властвуй». Он разделяет массив на более мелкие части, сортирует их и затем объединяет обратно в отсортированный массив.

Принцип работы:


1.   Разделение: массив рекурсивно делится пополам до тех пор, пока не останутся массивы размером в один элемент
2.   Сортировка: каждый подмассив сортируется отдельно
3.   Слияние: отсортированные подмассивы объединяются в один общий отсортированный массив


````python
def merge(left_list, right_list):
    sorted_list = []
    left_list_index = right_list_index = 0

    # Длины списков для удобства
    left_list_length, right_list_length = len(left_list), len(right_list)

    # Сравниваем элементы обоих списков и добавляем меньший в результирующий
    for _ in range(left_list_length + right_list_length):
        if left_list_index < left_list_length and right_list_index < right_list_length:
            if left_list[left_list_index] <= right_list[right_list_index]:
                sorted_list.append(left_list[left_list_index])
                left_list_index += 1
            else:
                sorted_list.append(right_list[right_list_index])
                right_list_index += 1
        # Если один из списков закончился, добавляем оставшиеся элементы
        elif left_list_index == left_list_length:
            sorted_list.append(right_list[right_list_index])
            right_list_index += 1
        elif right_list_index == right_list_length:
            sorted_list.append(left_list[left_list_index])
            left_list_index += 1
    return sorted_list

def merge_sort(nums):
    # Базовый случай: список из одного элемента уже отсортирован
    if len(nums) <= 1:
        return nums

    # Находим середину списка
    mid = len(nums) // 2

    # Рекурсивно сортируем левую и правую части
    left_list = merge_sort(nums[:mid])
    right_list = merge_sort(nums[mid:])

    # Объединяем отсортированные части
    return merge(left_list, right_list)

# Пример использования
random_list = [120, 45, 68, 250, 176]
sorted_list = merge_sort(random_list)
print(sorted_list)  # Вывод: [45, 68, 120, 176, 250]
````

Преимущества:

*   Стабильность сортировки
*   Гарантированная сложность O(n log n)
*   Хорошо работает с большими наборами данных

Недостатки:

*   Требует дополнительной памяти
*   Менее эффективен для небольших массивов по сравнению с другими алгоритмами

**Быстрая сортировка (Quick Sort)** — один из самых эффективных алгоритмов сортировки, основанный на принципе “разделяй и властвуй”. Он особенно эффективен при работе с большими наборами данных.

Принцип работы:


1.   **Выбор опорного элемента (pivot)** — элемента, относительно которого будет происходить разделение массива
2.   **Разделение массива** — перестановка элементов так, чтобы слева от опорного были меньшие элементы, а справа — большие
3.   **Рекурсивная сортировка** подмассивов слева и справа от опорного элемента



````python
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

# Пример использования
arr = [10, 7, 8, 9, 1, 5]
quick_sort(arr, 0, len(arr) - 1)
print("Отсортированный массив:", arr)
````

Характеристики алгоритма:


*   Временная сложность: В среднем: O(n log n). В худшем случае: O(n²) (когда опорный элемент всегда минимальный или максимальный)
*   Пространственная сложность: O(log n) из-за рекурсии
*   Устойчивость: алгоритм неустойчив

Важные моменты:


*   Выбор опорного элемента влияет на производительность
*   Алгоритм работает быстрее встроенных методов сортировки для больших наборов данных
*   Требует меньше памяти по сравнению с сортировкой слиянием
*   Эффективность падает при почти отсортированных массивах

Быстрая сортировка является отличным выбором для сортировки больших массивов данных благодаря своей высокой производительности и относительно простой реализации.

1.   Сортировка чисел по убыванию

*   Задача: отсортировать список чисел в порядке убывания, не используя встроенные методы сортировки.
*   Пример: `[5, 2, 9, 1, 5]` → `[9, 5, 5, 2, 1]`

````python
mass = [1, 2, 3, 4, 9, 32, 14, 56, 12]

def heap_create(mass, n, i):
    big = i
    l = 2 * i + 1
    r = 2 * i + 2
    
    if l < n and mass[i] < mass[l]:
        big = l
    
    if r < n and mass[big] < mass[r]:
        big = r
    
    if big != i:
        mass[i], mass[big] = mass[big],mass[i]  
        heap_create(mass, n, big)


# Пирамидальная сортировка
def heap_sort(mass):
    for i in range(len(mass)):
        n = len(mass)
        
        for i in range(n, -1, -1):
            heap_create(mass, n, i)
        
        for i in range(n - 1, 0, -1):
            mass[i], mass[0] = mass[0], mass[i]
            heap_create(mass, i, 0)
print('Изначальный массив', mass)
flag = input('По возрастанию? y/n ')
if flag == 'n':
    print('По убыванию')
    mass  = [-x for x in mass]
else:
    print('По возрастанию')
heap_sort(mass)
if flag == 'n':
    mass  = [-x for x in mass]
print('Отсортированыый массив', mass)
````

2.   Сортировка строк по длине


*   Задача: отсортировать список строк по их длине в порядке возрастания.
*   Пример: `["apple", "banana", "kiwi", "pear"]` → `["kiwi", "pear", "apple", "banana"]`

````python
mass = ["apple", "banana", "kiwi", "pear"]
print('Изначальный массив', mass)
flag = input('По возрастанию? y/n ')
if flag == 'n':
    print('По убыванию')
    mass.sort(key=lambda x: -len(x))
else:
    print('По возрастанию')
    mass.sort(key=lambda x: len(x))
print('Отсортированыый массив', mass)
````

3.   Сортировка по сумме цифр

*   Задача: отсортировать список чисел по сумме их цифр.
*   Пример: `[123, 45, 6, 789]` → `[6, 123, 45, 789]` (суммы: 6, 6, 6, 24)

````python
mass = [123, 45, 6, 789]
print('Изначальный массив', mass)
for i in range(len(mass)):
    mass[i] = [str(mass[i]), 0]
    j = 10
    for j in range(len(mass[i][0])):
        if mass[i][0][j] != ',' or mass[i][0][j] != '.':
            mass[i][1] += int(mass[i][0][j])
print(mass)
flag = input('По возрастанию? y/n ')
if flag == 'n':
    print('По убыванию')
    mass.sort(key=lambda x: -x[1])
else:
    print('По возрастанию')
    mass.sort(key=lambda x: x[1])
mass = [float(x[0]) for x in mass]
print('Отсортированыый массив', mass)
````

4.   Сортировка по частоте встречаемости

*   Задача: отсортировать список слов по частоте их встречаемости в тексте.
*   Пример: `["apple", "banana", "apple", "cherry", "banana"]` → `["apple", "banana", "cherry"]`

````python
mass = ["apple", "banana", "apple", "cherry", "banana"]
print('Изначальный массив', mass)
flag = input('По возрастанию? y/n ')
if flag == 'n':
    print('По убыванию')
    freq_dict = {x: -mass.count(x) for x in mass}
else:
    print('По возрастанию')
    freq_dict = {x: mass.count(x) for x in mass}
mass = sorted(freq_dict, key=lambda x: (freq_dict[x], x))
print('Отсортированыый массив', mass)
````

5.   Сортировка с сохранением исходных индексов

*   Задача: отсортировать список чисел, но сохранить исходные индексы элементов.
*   Пример: `[5, 2, 9, 1]` → `[(3, 1), (1, 2), (0, 5), (2, 9)]`

````python
mass = [5, 2, 9, 1]
print('Изначальный массив', mass)
mass = [(x, mass[x]) for x in range(len(mass))]
print('Массив с индексами', mass)
flag = input('По возрастанию? y/n ')
if flag == 'n':
    print('По убыванию')
    mass.sort(key=lambda x: x[1])
else:
    print('По возрастанию')
    mass.sort(key=lambda x: -x[1])
print('Отсортированыый массив', mass)
````

6.   Сортировка с удалением дубликатов

*   Задача: отсортировать список чисел, удалив дубликаты.
*   Пример: `[5, 2, 9, 1, 5, 2]` → `[1, 2, 5, 9]`

````python
mass = [5, 2, 9, 1, 5, 2]
print('Изначальный массив', mass)
mass = [x for x in set(mass)]
flag = input('По возрастанию? y/n ')
if flag == 'n':
    print('По убыванию')
    mass.sort(reverse=True)
else:
    print('По возрастанию')
    mass.sort()
print('Отсортированыый массив', mass)
````

4.   Несколько графиков

*   Построить на одном графике функции синуса и косинуса
*   Использовать разные цвета и стили линий

````python
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()

ax.plot(x, y1, label='sin(x)')
ax.plot(x, y2, label='cos(x)')

ax.legend()
plt.title('Графики $y = sin(x)$ и $y = cos(x)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()

plt.show();
````
